## Tuples

A tuple is a **fixed-length, immutable** sequence of python objects. The size and length and order and contents cannot be modified

```python
# tuple can be created with commas
tup = 4,5,6
tup
```

```python
# You can create a tuple of tuples, nested tuples, using parentheses
nested_tup = (4,5,6), (7,8)
nested_tup
```

```python
#convert a sequence or iterator to a tuple
tuple([4,0,2])

tupl = tuple("string")
tupl
```

```python
# to access objects in a tuple, use the square brackets
tupl[0]
```

The objects in a tuple themselves are mutable but the tuple itself is not mutable. Which means the order that the objects are in, can't be moved around.

```python
new_tup = ("rome", [4,5,6], True)
new_tup[1].append(3) #mutating the 2nd object in the tuple
new_tup
```

```python
# concatenating tuples
new_tup = new_tup + (1,2,3) + ("florence",)
new_tup
```

Tuples can be **unpacked**

```python
a, b, c = tup
b
```

```python
# unpacking nested tuples
tup1 = 4, 5, (6, 7)
a, b, (c, d) = tup1

c
```

```python
# Using *rest to unpack a tuple
values = 1,2,3,4,5
a, b, *rest = values #rest takes everything apart from the objects assigned to variables a and b
rest
```

Most people use the rest syntax to pluck objects that are to be discarded

Instances for a tuple:
- count

```python
a = (1,2,2,2,3,4,2)
a.count(2)
```

## List

Lists are **variale-length and their contents can be modified in-place**.  
They are contrasted to tuples

```python
# A list is defined with square brackets or the list type function
a_list = [2,3,7, None]
tup = ("foo", "bar", "baz")

b_list = list(tup)

b_list
```

```python
#Changing the objects in a list
b_list[1] = "rome"
b_list
```

```python
# Lists are used to materialise an iterator
gen = range(10)
gen
```

```python
list(gen)
```

**Adding and removing elements in a list**

```python
#Adding elements to the end of a list
#Append
b_list.append("dwarf")
b_list
```

```python
#Adding element into a specific location in a list
#Insert
b_list.insert(1, "red")
b_list
```

Insert is computationally expensive though

```python
#The inverse of insert. Removing an item from a specific location
#pop
b_list.pop(2)
b_list
```

```python
# Elements can be removed by looking for the first of the value passed
#remove
b_list.remove("red")
b_list
```

Checking if a list has a value is a lot slower computationally than doing so with **dicts and sets**. This is because Python makes a linear scan across the value of the entire list.

```python
"dwarf" in b_list
```

```python
"dwarf" not in b_list
```

**Concatenating and combining lists**

```python
#Use + to concat
concat = [4, None, "foo"] + [7,8,(2,3)]
concat
```

```python
#Append multiple items to the end of a list
#Extend
x = [4, None, "foo"]

x.extend([7,8, (2,3)])

x
```

Using addition to concatenate is more computationally expensive than using extend. This is because addition creates a new list and then adds them together, extend, on the other hand, just appends it to the end of the list.

## Python built-in sequence functions

```python
#Enumerate
#If you want the value and the position of every item in a loop

fruits = ["apple", "bananas", "orange"]

for i, fruit in enumerate(fruits):
    print(i, fruit)
```

```python
#Sorted
#sorts and returns the newly sorted list

sorted(c)
```

```python
#Zip
#zip pairs elements in x number of sequences to create a list of tuples

seq1 = ["foo", "bar", "baz"]
seq2 = ["one", "two", "three"]

zipped = zip(seq1, seq2)

list(zipped)
```

```python
# if a sequence is shorter than the other, zip stops where they all meet
seq3 = ["True", "False"]

zipper = zip(seq1, seq2, seq3)

list(zipper)
#there is no three or baz in the results
```

```python
#Zip stops you having to index between multiple lists
names = ["Alice", "Bob", "Charlie"]
ages = [25, 30, 35]

for name, age in zip(names, ages):
    print(name, age)
```

```python
#Zip allows you to unzip a sequence also
pitchers = [("Nolan", "Ryan"), ("Roger", "Clemens"), ("Schilling", "Curt")]

first_name, last_name = zip(*pitchers) #the * does the unzipping

first_name
```

```python
#Reversed
#Iterates over a sequence in reverse order
#is a generator so need to use lit or a for loop to materialise it

list(reversed(range(10)))
```

## Dictionary

Also called a hash map or associative array.  
Has key-value pairs and the key and values are Python objects.  
  
A dict is essentially a collection of 2 tuples

```python
empty_dict = {} #curly braces to start a dictionary

d1 = {"a" : "some value", "b" : "[1,2,3,4]"}

d1
```

Access, insert and set elements, use the same syntax as a list or tuple

```python
d1[7] = "integer"
d1
```

```python
d1["b"]
```

Can check if a **key** is in a dict using the same syntax for a list or tuple

```python
"b" in d1
```

**Deleting values**

```python
#del method
d1[5] = "random value"
d1
```

```python
del d1[5]
d1
```

```python
#pop method
ret = d1.pop(7)
ret
```

**Adding to a dict**

```python
#update
#update also changes their values in-place

d1.update({"c" : "foo", "d" : 12})

d1
```

**Dicts from sequences**

```python
mapping = dict(zip(range(5), reversed(range(5))))

mapping
```

**Retrieving values from a key search**

```python
#Get
# Use get instead of a normal search

d1.get("c", "car") #get the value of c and if there is none, make it car and return it
```

```python
#what if there is no key or value with that name
#It returns nothing

#this is better than using d1["e"] which would return a KeyError

d1.get("e")
```

```python
#SetDefault
#dict.setdefault(key, default)
#Returns the value if it is present, if not, creates the jey with the default and returns it

#Best use is when creating dictionaries of lists, sets etc

words = ["apple", "atom", "bat", "bar", "book"]

by_letter = {}

for word in words:
    letter = word[0]
    by_letter.setdefault(letter, []).append(word)
```

```python
by_letter
```

```python
#Automatically creates an empty list whenever a missing key is accessed
#defaultdict(list)
#best way to group or accumulate values

from collections import defaultdict
by_letters = defaultdict(list)
for word in words:
    by_letters[word[0]].append(word)

by_letters
```

What can be used as a value in a dict?  
It has to be immutable, such as
- has to be a scalar type (int, float, string)
- tuple

It is essentially about hashability

```python
#can check if someting is hashable
hash("string")
```

```python
#fails because a list is mutable
hash([1,2,3])
```

```python
#How to use a list as a key
#convert it to a tuple

d = {}

d[tuple([1,2,3])]=5

d
```

## List,set & dict comprehensions

```python
#List comprehension
# [Expression for var in collection if condition]

strings = ["a", "as", "bat", "car", "dove", "python"]

[x.upper() for x in strings if len(x)>2]
```

```python
#dict comprehension
# dict_comp = {key-expression : value-expression for value in collection if condition}

loc_mapping  = {val:index for index, val in enumerate(strings)}
loc_mapping
```

```python
#set comprehension
# set_comp = {expression for value in collection if condition}

unique_lengths = {len(x) for x in strings}

unique_lengths
```
