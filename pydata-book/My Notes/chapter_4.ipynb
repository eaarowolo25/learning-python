## NumPy ndarray - multidimensional array object

ndarray has:
- all elements are of the same type
- every array has a shape - size of each dimension of the array
- every array has a dtype - the data type of the arry

```python
import numpy as np
```

```python
data = np.random.randn(2,3)
data
```

```python
#shape of the array
data.shape
```

```python
#data type of the array
data.dtype
```

**Creating ndarrays**

```python
# array function
# accepts any sequence-like object and produces a new NumPy array

data1 = [6, 7.5, 8, 0, 1]

arr1 = np.array(data1)

arr1
```

```python
# nested sequences like a list of equal-length lists will become a multidimensional array

data2 = [[1,2,3,4], [5,6,7,8]]

arr2 = np.array(data2)

arr2
```

```python
#finding out the dimensions of an array
# use ndim
arr2.ndim
```

```python
#finding out the shape of an array
#use shape

arr2.shape
```

The functions zeros, ones and empty also make arrays. You just have to pass the dimension to it

```python
#zeros
np.zeros(10)
```

```python
#empty
np.empty((2,3,2))
```

```python
#Numpy has a built in function similar to the python range function
#arange

np.arange(15)
```
