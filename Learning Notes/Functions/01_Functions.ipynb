# Functions

You can assign a function to a variable and use that variable the same way you would call a function

```python
def func():
    print("I am function func()!")

func()
```

```python
another_name = func

another_name()
```

A function can be: displayed with print(), included as an element in a list, and used in a dictionary as a key

```python
#function in print()
print("cat", func, 42)

#function in a list
objects = ["cat", func, 42]
print(objects[1]())

#function in a dictionary
d = {"cat":1, func: 2, 42:3}
d[func]
```

### *Function Composition*  
This is when you pass a function to another function as an argument

```python
def inner():
    print("I am function inner()")

def outer(function):
    function()

outer(inner)
```

An example of this is the built in function sorted(). You can give it an optional key argument that serves as the callback function.

```python
animals = ["ferret", "vole", "dog", "gecko"]

# Sorted by itself
print("No key sorted:", sorted(animals))

#Sorted with length key
print("Sorted with length key:", sorted(animals, key=len))

#sorted with length and reverse
print("Sorted with length and reverse:", sorted(animals, key=len, reverse=True))
```

```python
#now we can pass a function for the key
def reverse_len(s):
    return -len(s)

#instead of reverse and len, we just put them together
sorted(animals,  key=reverse_len)
```

A function can also specify another function as its return value

```python
def outer():
    def inner():
        print("I am function inner()!!")
    #function outer returns function inner
    return inner

#assigning the return value of outer() to the variable function
function = outer()
function

function()
outer()()
```
