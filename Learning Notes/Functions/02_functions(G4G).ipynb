## Function arguments

Arguments are values passed to a function when it is called. This allows functions to get the input data and perform repeatable operations on those values.

```python
"""
def function_name(arguments):
    #function body
    return value
"""
```

```python
#Example
#function checks if a number is even or odd

def evenodd(x):
    if (x%2 == 0):
        return "Even"
    else:
        return "Odd"
    
print(evenodd(9))
print(evenodd(12))
```

### Types of Function Arguments

#### Default Argument

Default arguments use a predefined value when no value is passed during the function call.

```python
def myfunc(x, y=50):
    print("x: ", x)
    print("y: ", y)

myfunc(10)
#Since y isnt provided, it uses the default value
```

#### Keyword Arguments

pass values using parameter names

```python
def student(fname, lname):
    print(fname, lname)

#The order that arguments are passed do not matter
student(fname="Romane", lname="Dogba")
student(lname="Arowolo", fname="Emmanuel")
```

#### Positional Arguments

values are assigned to parameters based on their order in the function call

```python
def nameage(name, age):
    print("Hi, I am", name)
    print("My age is", age)

print("Case1:")
nameage("Olivia", 27)

print("Case2:")
nameage(27, "olivia")
```

#### Arbritrary Arguments

Allow functions to accept multiple values. This is done using two special symbols:  
*args - collects an extra positional argument as a tuple  
**kwargs collects an extra keywaord as a dictionary

```python
def myfunc(*args, **kwargs):
    print("Non-Keyword Arguments (*args):")
    for arg in args:
        print(arg)
    
    print ("Keyword Arguments (**kwargs):")
    for key, value in kwargs.items():
        print(f"{key} == {value}")

myfunc("Hey", "welcome", first = "Emmanuel", mid="is", last="Arowolo")
```

## Function in a function

A function defined inside of a function is known as a **nested function**. It is used to organise related logic and access variables from the outer function.

```python
def f1():
    s = "I love functions"
    def f2():
        print(s)
    return f2()

f1()
```

## Return Statement

Return is used to end a function and send a value back to the caller. Values can be returned as any data type and if multiple values, it will be returned as a tuple.  
If no value is returned then it will return None by default

```python
def sq_value(num):
    return num**2

print(sq_value(2))
print(sq_value(3))
```

```python
# list is a mutable object
def myFun(x):
    x[0] = 20

b = [10, 11, 12, 13]
myFun(b)
print(b)

#integer is immutable
def myFun2(x):
    x = 20

a = 10

myFun2(a)
print(a)
```
